```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# ==================== –ê–í–¢–û–ú–ê–¢–ò–ß–ï–°–ö–ê–Ø –ó–ê–ì–†–£–ó–ö–ê –î–ê–ù–ù–´–• ====================
print("="*50)
print("–ê–ù–ê–õ–ò–ó –£–î–û–í–õ–ï–¢–í–û–†–ï–ù–ù–û–°–¢–ò –ü–ê–°–°–ê–ñ–ò–†–û–í –ê–í–ò–ê–ö–û–ú–ü–ê–ù–ò–ô")
print("="*50)

def load_airline_data():
    """–ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –∑–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤"""
    print("–ü–æ–ø—ã—Ç–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –¥–∞–Ω–Ω—ã—Ö –æ–± –∞–≤–∏–∞–ø–∞—Å—Å–∞–∂–∏—Ä–∞—Ö...")

    # –°–ø–æ—Å–æ–± 1: –° GitHub (–æ—Å–Ω–æ–≤–Ω–æ–π)
    try:
        print("1. –ó–∞–≥—Ä—É–∑–∫–∞ —Å GitHub...")
        url = "https://raw.githubusercontent.com/annytab/Data/main/Airline%20Passenger%20Satisfaction.csv"
        df = pd.read_csv(url)
        print("   ‚úì –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ —Å GitHub")
        return df
    except Exception as e:
        print(f"   ‚úó –û—à–∏–±–∫–∞: {e}")

    # –°–ø–æ—Å–æ–± 2: –õ–æ–∫–∞–ª—å–Ω—ã–π —Ñ–∞–π–ª
    try:
        print("2. –ü–æ–∏—Å–∫ –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ —Ñ–∞–π–ª–∞...")
        df = pd.read_csv('airline_passenger_satisfaction.csv')
        print("   ‚úì –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ –∏–∑ –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ —Ñ–∞–π–ª–∞")
        return df
    except:
        try:
            df = pd.read_csv('train.csv')
            print("   ‚úì –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ –∏–∑ train.csv")
            return df
        except Exception as e:
            print(f"   ‚úó –û—à–∏–±–∫–∞: {e}")

    # –°–ø–æ—Å–æ–± 3: –°–æ–∑–¥–∞–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    print("3. –°–æ–∑–¥–∞–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö...")
    np.random.seed(42)
    n_samples = 1000

    data = {
        'Gender': np.random.choice(['Male', 'Female'], n_samples),
        'Customer Type': np.random.choice(['Loyal Customer', 'disloyal Customer'], n_samples),
        'Age': np.random.randint(18, 70, n_samples),
        'Type of Travel': np.random.choice(['Business travel', 'Personal Travel'], n_samples),
        'Class': np.random.choice(['Business', 'Eco', 'Eco Plus'], n_samples),
        'Flight Distance': np.random.exponential(1000, n_samples).astype(int),
        'Departure Delay in Minutes': np.random.exponential(20, n_samples).astype(int),
        'Arrival Delay in Minutes': np.random.exponential(20, n_samples).astype(int),
        'satisfaction': np.random.choice(['satisfied', 'neutral or dissatisfied'], n_samples)
    }

    # –†–µ–π—Ç–∏–Ω–≥–∏ 0-5
    for i in range(1, 24):
        data[f'Inflight wifi service'] = np.random.randint(0, 6, n_samples)
        data[f'Food and drink'] = np.random.randint(0, 6, n_samples)

    df = pd.DataFrame(data)
    print("   ‚úì –°–æ–∑–¥–∞–Ω—ã —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ")
    return df

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
df = load_airline_data()
print(f"\n‚úÖ –î–∞–Ω–Ω—ã–µ —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
print(f"üìä –†–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö: {df.shape}")
print(f"üìã –ö–æ–ª–æ–Ω–∫–∏ ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:2d}. {col}")

# ==================== A. –ó–ê–ì–†–£–ó–ö–ê –ò –ü–ï–†–í–ò–ß–ù–´–ô –û–ë–ó–û–† ====================
print("\n" + "="*50)
print("A. –ó–ê–ì–†–£–ó–ö–ê –ò –ü–ï–†–í–ò–ß–ù–´–ô –û–ë–ó–û–†")
print("="*50)

# –ü–µ—Ä–≤—ã–µ –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏
print("\n1. –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫:")
print(df.head())
print("\n–ü–æ—Å–ª–µ–¥–Ω–∏–µ 5 —Å—Ç—Ä–æ–∫:")
print(df.tail())

# –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
print(f"\n2. –†–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö: {df.shape[0]} —Å—Ç—Ä–æ–∫, {df.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤")

# –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö
print("\n3. –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:")
print(df.dtypes.value_counts())
print("\n–î–µ—Ç–∞–ª—å–Ω–æ:")
for col in df.columns:
    print(f"  {col:30} : {str(df[col].dtype):10} | –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö: {df[col].nunique()}")

# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤
print("\n4. –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö...")
df_clean = df.copy()

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–∏—Å–ª–æ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
print(f"  –ß–∏—Å–ª–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏ ({len(numeric_cols)}): {numeric_cols}")

# –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏ (–Ω–µ—á–∏—Å–ª–æ–≤—ã–µ –∏ —Å –º–∞–ª—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π)
cat_cols = []
for col in df_clean.columns:
    if col not in numeric_cols or df_clean[col].nunique() < 10:
        if df_clean[col].nunique() < 50:  # –ù–µ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –∫–æ–ª–æ–Ω–∫–∏ —Å–æ –º–Ω–æ–≥–∏–º–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
            df_clean[col] = df_clean[col].astype('category')
            cat_cols.append(col)

print(f"  –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏ ({len(cat_cols)}): {cat_cols[:10]}...")

# ==================== B. –û–ë–†–ê–ë–û–¢–ö–ê –î–ê–ù–ù–´–• ====================
print("\n" + "="*50)
print("B. –û–ë–†–ê–ë–û–¢–ö–ê –î–ê–ù–ù–´–•")
print("="*50)

# 1. –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
print("\n1. –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
missing_before = df_clean.isnull().sum()
if missing_before.sum() > 0:
    print("–ü—Ä–æ–ø—É—Å–∫–∏ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º:")
    for col, count in missing_before[missing_before > 0].items():
        percentage = (count / len(df_clean)) * 100
        print(f"  {col:30} : {count:4} –ø—Ä–æ–ø—É—Å–∫–æ–≤ ({percentage:.1f}%)")

    # –°—Ç—Ä–∞—Ç–µ–≥–∏—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è
    print("\n–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤...")
    for col in df_clean.columns:
        if df_clean[col].isnull().sum() > 0:
            if df_clean[col].dtype in ['int64', 'float64']:
                # –î–ª—è —á–∏—Å–ª–æ–≤—ã—Ö - –º–µ–¥–∏–∞–Ω–∞
                fill_value = df_clean[col].median()
                df_clean[col] = df_clean[col].fillna(fill_value)
                print(f"  {col:30} : –∑–∞–ø–æ–ª–Ω–µ–Ω–æ –º–µ–¥–∏–∞–Ω–æ–π = {fill_value}")
            else:
                # –î–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö - –º–æ–¥–∞
                fill_value = df_clean[col].mode()[0] if len(df_clean[col].mode()) > 0 else 'Unknown'
                df_clean[col] = df_clean[col].fillna(fill_value)
                print(f"  {col:30} : –∑–∞–ø–æ–ª–Ω–µ–Ω–æ –º–æ–¥–æ–π = '{fill_value}'")
else:
    print("  ‚úì –ü—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ")

print(f"\n–ü—Ä–æ–ø—É—Å–∫–∏ –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏: {df_clean.isnull().sum().sum()}")

# 2. –î—É–±–ª–∏–∫–∞—Ç—ã
print("\n2. –î—É–±–ª–∏–∫–∞—Ç—ã:")
duplicates = df_clean.duplicated().sum()
print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {duplicates}")
if duplicates > 0:
    df_clean = df_clean.drop_duplicates()
    print(f"  –£–¥–∞–ª–µ–Ω–æ {duplicates} –¥—É–±–ª–∏–∫–∞—Ç–æ–≤")
    print(f"  –ù–æ–≤—ã–π —Ä–∞–∑–º–µ—Ä: {df_clean.shape}")
else:
    print("  ‚úì –î—É–±–ª–∏–∫–∞—Ç–æ–≤ –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ")

# 3. –í—ã–±—Ä–æ—Å—ã
print("\n3. –ê–Ω–∞–ª–∏–∑ –≤—ã–±—Ä–æ—Å–æ–≤:")
if numeric_cols:
    print("  –ê–Ω–∞–ª–∏–∑ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤:")
    outlier_info = []

    for col in numeric_cols:
        if df_clean[col].dtype in ['int64', 'float64']:
            # –ú–µ—Ç–æ–¥ IQR
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
            outlier_count = len(outliers)

            if outlier_count > 0:
                outlier_percentage = (outlier_count / len(df_clean)) * 100
                outlier_info.append((col, outlier_count, outlier_percentage, lower_bound, upper_bound))

    if outlier_info:
        print("  –û–±–Ω–∞—Ä—É–∂–µ–Ω—ã –≤—ã–±—Ä–æ—Å—ã:")
        print("  –ö–æ–ª–æ–Ω–∫–∞                    | –í—ã–±—Ä–æ—Å–æ–≤ | %     | –ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ | –í–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞")
        print("  " + "-"*80)
        for col, count, pct, low, high in outlier_info:
            print(f"  {col:25} | {count:8} | {pct:5.1f} | {low:13.2f} | {high:13.2f}")
    else:
        print("  ‚úì –í—ã–±—Ä–æ—Å–æ–≤ –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ")
else:
    print("  –ù–µ—Ç —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤—ã–±—Ä–æ—Å–æ–≤")

# 4. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–∞–Ω–Ω—ã—Ö
print("\n4. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–∞–Ω–Ω—ã—Ö:")
print("  –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
print(df_clean.describe().round(2))

print("\n  –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏:")
cat_cols = df_clean.select_dtypes(include=['category', 'object']).columns
for col in cat_cols[:5]:  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 5
    print(f"\n  {col}:")
    value_counts = df_clean[col].value_counts()
    for val, count in value_counts.head(3).items():
        percentage = (count / len(df_clean)) * 100
        print(f"    '{val}': {count} ({percentage:.1f}%)")
    if len(value_counts) > 3:
        print(f"    ... –∏ –µ—â–µ {len(value_counts) - 3} –∑–Ω–∞—á–µ–Ω–∏–π")

# –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑
print("\n  –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
numeric_df = df_clean.select_dtypes(include=[np.number])
if len(numeric_df.columns) > 1:
    correlation = numeric_df.corr()
    print("  –°–∞–º—ã–µ —Å–∏–ª—å–Ω—ã–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:")

    # –ù–∞—Ö–æ–¥–∏–º –ø–∞—Ä—ã —Å –≤—ã—Å–æ–∫–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–µ–π
    strong_corr = []
    for i in range(len(correlation.columns)):
        for j in range(i+1, len(correlation.columns)):
            corr_value = abs(correlation.iloc[i, j])
            if corr_value > 0.7:  # –ü–æ—Ä–æ–≥ —Å–∏–ª—å–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
                strong_corr.append((correlation.columns[i], correlation.columns[j], corr_value))

    if strong_corr:
        for col1, col2, corr in sorted(strong_corr, key=lambda x: x[2], reverse=True):
            print(f"    {col1} ‚Üî {col2}: {corr:.3f}")
    else:
        print("    ‚úì –°–∏–ª—å–Ω—ã—Ö –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ")
else:
    print("  –ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞")

# ==================== C. –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø ====================
print("\n" + "="*50)
print("C. –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –î–ê–ù–ù–´–•")
print("="*50)
print("–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤...")

plt.figure(figsize=(20, 16))
sns.set_style("whitegrid")
sns.set_palette("husl")

# 1. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏
plt.subplot(3, 3, 1)
if 'satisfaction' in df_clean.columns:
    satisfaction_counts = df_clean['satisfaction'].value_counts()
    colors = ['#4CAF50', '#F44336'] if len(satisfaction_counts) == 2 else None
    satisfaction_counts.plot(kind='bar', color=colors, edgecolor='black', alpha=0.8)
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤', fontsize=14, fontweight='bold')
    plt.xlabel('–£—Ä–æ–≤–µ–Ω—å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏', fontsize=12)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤', fontsize=12)
    plt.xticks(rotation=45)

    for i, v in enumerate(satisfaction_counts.values):
        plt.text(i, v + max(satisfaction_counts.values)*0.01, str(v),
                ha='center', fontweight='bold')

# 2. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–ª–∞—Å—Å–∞–º
plt.subplot(3, 3, 2)
if 'Class' in df_clean.columns:
    class_counts = df_clean['Class'].value_counts()
    colors = ['gold', 'silver', '#CD7F32'][:len(class_counts)]
    class_counts.plot(kind='bar', color=colors, edgecolor='black', alpha=0.8)
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤ –ø–æ –∫–ª–∞—Å—Å–∞–º', fontsize=14, fontweight='bold')
    plt.xlabel('–ö–ª–∞—Å—Å –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è', fontsize=12)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ', fontsize=12)

    for i, v in enumerate(class_counts.values):
        plt.text(i, v + max(class_counts.values)*0.01, str(v),
                ha='center', fontweight='bold')

# 3. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞
plt.subplot(3, 3, 3)
if 'Age' in df_clean.columns:
    df_clean['Age'].hist(bins=30, edgecolor='black', color='skyblue', alpha=0.7)
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤', fontsize=14, fontweight='bold')
    plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç', fontsize=12)
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

    mean_age = df_clean['Age'].mean()
    median_age = df_clean['Age'].median()
    plt.axvline(mean_age, color='red', linestyle='--', label=f'–°—Ä–µ–¥–Ω–µ–µ: {mean_age:.1f}')
    plt.axvline(median_age, color='green', linestyle='-.', label=f'–ú–µ–¥–∏–∞–Ω–∞: {median_age:.1f}')
    plt.legend()

# 4. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –ø–æ–ª–µ—Ç–∞
plt.subplot(3, 3, 4)
if 'Flight Distance' in df_clean.columns:
    df_clean['Flight Distance'].hist(bins=50, edgecolor='black', color='lightgreen', alpha=0.7)
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –ø–æ–ª–µ—Ç–∞', fontsize=14, fontweight='bold')
    plt.xlabel('–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ (–∫–º)', fontsize=12)
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

    mean_dist = df_clean['Flight Distance'].mean()
    plt.axvline(mean_dist, color='red', linestyle='--',
                label=f'–°—Ä–µ–¥–Ω–µ–µ: {mean_dist:.0f} –∫–º')

# 5. –ó–∞–¥–µ—Ä–∂–∫–∏ –≤—ã–ª–µ—Ç–∞
plt.subplot(3, 3, 5)
if 'Departure Delay in Minutes' in df_clean.columns:
    delay_data = df_clean['Departure Delay in Minutes']
    # –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è —à–∫–∞–ª–∞ –¥–ª—è –ª—É—á—à–µ–π –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
    log_delay = np.log1p(delay_data)
    plt.hist(log_delay, bins=40, edgecolor='black', color='orange', alpha=0.7)
    plt.title('–ó–∞–¥–µ—Ä–∂–∫–∏ –≤—ã–ª–µ—Ç–∞ (–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è —à–∫–∞–ª–∞)', fontsize=14, fontweight='bold')
    plt.xlabel('ln(–ó–∞–¥–µ—Ä–∂–∫–∞ + 1)', fontsize=12)
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

# 6. Heatmap –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
plt.subplot(3, 3, 6)
if len(numeric_df.columns) > 1:
    corr_matrix = numeric_df.corr()
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=False, cmap='coolwarm',
                center=0, square=True, cbar_kws={'shrink': 0.8})
    plt.title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞', fontsize=14, fontweight='bold')
else:
    plt.text(0.5, 0.5, '–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö\n–¥–ª—è heatmap',
             ha='center', va='center', fontsize=12)

# 7. –£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø–æ –∫–ª–∞—Å—Å–∞–º
plt.subplot(3, 3, 7)
if all(col in df_clean.columns for col in ['satisfaction', 'Class']):
    satisfaction_by_class = pd.crosstab(df_clean['Class'], df_clean['satisfaction'], normalize='index')
    satisfaction_by_class.plot(kind='bar', stacked=True, edgecolor='black')
    plt.title('–£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø–æ –∫–ª–∞—Å—Å–∞–º', fontsize=14, fontweight='bold')
    plt.xlabel('–ö–ª–∞—Å—Å', fontsize=12)
    plt.ylabel('–î–æ–ª—è', fontsize=12)
    plt.legend(title='–£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å')

# 8. Boxplot –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–æ –∫–ª–∞—Å—Å–∞–º
plt.subplot(3, 3, 8)
if all(col in df_clean.columns for col in ['Age', 'Class']):
    sns.boxplot(x='Class', y='Age', data=df_clean)
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–æ –∫–ª–∞—Å—Å–∞–º', fontsize=14, fontweight='bold')
    plt.xlabel('–ö–ª–∞—Å—Å', fontsize=12)
    plt.ylabel('–í–æ–∑—Ä–∞—Å—Ç', fontsize=12)

# 9. –¢–∏–ø –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏—è
plt.subplot(3, 3, 9)
if 'Type of Travel' in df_clean.columns:
    travel_counts = df_clean['Type of Travel'].value_counts()
    plt.pie(travel_counts.values, labels=travel_counts.index, autopct='%1.1f%%',
            startangle=90, colors=['lightblue', 'lightcoral'])
    plt.title('–¢–∏–ø –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏—è', fontsize=14, fontweight='bold')

plt.suptitle('–ê–ù–ê–õ–ò–ó –£–î–û–í–õ–ï–¢–í–û–†–ï–ù–ù–û–°–¢–ò –ê–í–ò–ê–ü–ê–°–°–ê–ñ–ò–†–û–í', fontsize=18, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ==================== D. –ì–†–£–ü–ü–ò–†–û–í–ö–ò –ò –ê–ì–†–ï–ì–ê–¶–ò–ò ====================
print("\n" + "="*50)
print("D. –ì–†–£–ü–ü–ò–†–û–í–ö–ò –ò –ê–ì–†–ï–ì–ê–¶–ò–ò")
print("="*50)

# 1. –û—Å–Ω–æ–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º
print("\n1. –û—Å–Ω–æ–≤–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø–æ –∫–ª–∞—Å—Å–∞–º –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è:")
if 'Class' in df_clean.columns:
    group_cols = {}
    if 'Age' in df_clean.columns:
        group_cols['Age'] = ['mean', 'median', 'std']
    if 'Flight Distance' in df_clean.columns:
        group_cols['Flight Distance'] = ['mean', 'median']
    if 'Departure Delay in Minutes' in df_clean.columns:
        group_cols['Departure Delay in Minutes'] = ['mean', 'median']

    if group_cols:
        group1 = df_clean.groupby('Class').agg(group_cols).round(2)
        group1['Passenger Count'] = df_clean.groupby('Class').size()
        print(group1)
    else:
        print("  –ù–µ—Ç —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∞–≥—Ä–µ–≥–∞—Ü–∏–∏")

# 2. –£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø–æ —Ç–∏–ø–∞–º –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏—è
print("\n2. –£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø–æ —Ç–∏–ø—É –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏—è:")
if all(col in df_clean.columns for col in ['Type of Travel', 'satisfaction']):
    satisfaction_matrix = pd.crosstab(df_clean['Type of Travel'],
                                     df_clean['satisfaction'],
                                     margins=True)
    satisfaction_matrix['Satisfaction Rate'] = (
        satisfaction_matrix.iloc[:, 0] / satisfaction_matrix['All'] * 100
    ).round(1)
    print(satisfaction_matrix)

# 3. –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞: –≤–æ–∑—Ä–∞—Å—Ç –∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ –∫–ª–∞—Å—Å—É –∏ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏
print("\n3. –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ (–≤–æ–∑—Ä–∞—Å—Ç –∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ):")
if all(col in df_clean.columns for col in ['Class', 'satisfaction', 'Age', 'Flight Distance']):
    pivot_table = pd.pivot_table(
        df_clean,
        values=['Age', 'Flight Distance'],
        index='Class',
        columns='satisfaction',
        aggfunc=['mean', 'median', 'count']
    ).round(1)
    print(pivot_table)

# 4. –ó–∞–¥–µ—Ä–∂–∫–∏ –ø–æ –∫–ª–∞—Å—Å–∞–º
print("\n4. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∑–∞–¥–µ—Ä–∂–µ–∫ –ø–æ –∫–ª–∞—Å—Å–∞–º:")
if all(col in df_clean.columns for col in ['Class', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']):
    delay_stats = df_clean.groupby('Class').agg({
        'Departure Delay in Minutes': ['mean', 'median', 'max', 'std'],
        'Arrival Delay in Minutes': ['mean', 'median', 'max', 'std']
    }).round(1)
    print(delay_stats)

# 5. –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –ø–æ –ø–æ–ª—É –∏ –∫–ª–∞—Å—Å—É
print("\n5. –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –ø–æ –ø–æ–ª—É –∏ –∫–ª–∞—Å—Å—É:")
if all(col in df_clean.columns for col in ['Gender', 'Class', 'satisfaction']):
    detail_analysis = df_clean.groupby(['Gender', 'Class', 'satisfaction']).size().unstack()
    detail_analysis['Total'] = detail_analysis.sum(axis=1)
    detail_analysis['Satisfaction Rate'] = (
        detail_analysis.iloc[:, 0] / detail_analysis['Total'] * 100
    ).round(1)
    print(detail_analysis)

# 6. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –∑–∞–¥–µ—Ä–∂–µ–∫ —Å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å—é
print("\n6. –í–ª–∏—è–Ω–∏–µ –∑–∞–¥–µ—Ä–∂–µ–∫ –Ω–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å:")
if all(col in df_clean.columns for col in ['Departure Delay in Minutes', 'Arrival Delay in Minutes', 'satisfaction']):
    # –°–æ–∑–¥–∞–µ–º –±–∏–Ω–∞—Ä–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏
    df_clean['satisfaction_numeric'] = df_clean['satisfaction'].apply(
        lambda x: 1 if str(x).lower().startswith('satisf') else 0
    )

    delay_corr = df_clean[['Departure Delay in Minutes',
                          'Arrival Delay in Minutes',
                          'satisfaction_numeric']].corr()
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –∑–∞–¥–µ—Ä–∂–µ–∫ —Å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å—é:")
    print(delay_corr.round(3))

# ==================== –ò–¢–û–ì–ò ====================
print("\n" + "="*50)
print("–ò–¢–û–ì–û–í–ê–Ø –ò–ù–§–û–†–ú–ê–¶–ò–Ø")
print("="*50)

print(f"üìä –û–±—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏:")
print(f"  –í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {len(df_clean):,}")
print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {len(df_clean.columns)}")
print(f"  –ü—Ä–æ–ø—É—Å–∫–æ–≤: {df_clean.isnull().sum().sum()}")
print(f"  –î—É–±–ª–∏–∫–∞—Ç–æ–≤ –±—ã–ª–æ: {duplicates}")

# –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ –∫–ª—é—á–µ–≤—ã–º –∫–æ–ª–æ–Ω–∫–∞–º
key_stats = []
if 'Age' in df_clean.columns:
    key_stats.append(f"  –°—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç: {df_clean['Age'].mean():.1f} –ª–µ—Ç")
if 'Flight Distance' in df_clean.columns:
    key_stats.append(f"  –°—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ: {df_clean['Flight Distance'].mean():.0f} –∫–º")
if 'Departure Delay in Minutes' in df_clean.columns:
    key_stats.append(f"  –°—Ä–µ–¥–Ω—è—è –∑–∞–¥–µ—Ä–∂–∫–∞ –≤—ã–ª–µ—Ç–∞: {df_clean['Departure Delay in Minutes'].mean():.1f} –º–∏–Ω")

for stat in key_stats:
    print(stat)

# –£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å
if 'satisfaction' in df_clean.columns:
    if df_clean['satisfaction'].dtype.name == 'category':
        satisfaction_counts = df_clean['satisfaction'].value_counts()
        total = len(df_clean)
        for status, count in satisfaction_counts.items():
            percentage = (count / total) * 100
            print(f"  {status}: {count} ({percentage:.1f}%)")

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("\nüíæ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤...")
try:
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
    df_clean.to_csv('airline_satisfaction_processed.csv', index=False)
    print("  ‚úì –û–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ 'airline_satisfaction_processed.csv'")

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –æ—Ç—á–µ—Ç
    with open('airline_analysis_report.txt', 'w') as f:
        f.write("–û–¢–ß–ï–¢ –ü–û –ê–ù–ê–õ–ò–ó–£ –£–î–û–í–õ–ï–¢–í–û–†–ï–ù–ù–û–°–¢–ò –ê–í–ò–ê–ü–ê–°–°–ê–ñ–ò–†–û–í\n")
        f.write("="*50 + "\n")
        f.write(f"–í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {len(df_clean)}\n")
        f.write(f"–ö–æ–ª–æ–Ω–æ–∫: {len(df_clean.columns)}\n")
        if 'satisfaction' in df_clean.columns:
            f.write("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏:\n")
            for status, count in df_clean['satisfaction'].value_counts().items():
                pct = (count / len(df_clean)) * 100
                f.write(f"  {status}: {count} ({pct:.1f}%)\n")

    print("  ‚úì –¢–µ–∫—Å—Ç–æ–≤—ã–π –æ—Ç—á–µ—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω –≤ 'airline_analysis_report.txt'")

except Exception as e:
    print(f"  ‚úó –û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–∏: {e}")

print("\n" + "="*50)
print("–ê–ù–ê–õ–ò–ó –ó–ê–í–ï–†–®–ï–ù –£–°–ü–ï–®–ù–û! üéâ")
print("="*50)
print("\nüìã –í–´–ü–û–õ–ù–ï–ù–û:")
print("  ‚Ä¢ –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –∑–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö")
print("  ‚Ä¢ –û—á–∏—Å—Ç–∫–∞ –∏ –æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö")
print("  ‚Ä¢ –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ 9 –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–π")
print("  ‚Ä¢ –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ 6 –≥—Ä—É–ø–ø–∏—Ä–æ–≤–æ–∫ –∏ –∞–≥—Ä–µ–≥–∞—Ü–∏–π")
print("  ‚Ä¢ –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –∏—Ç–æ–≥–æ–≤–æ–≥–æ –æ—Ç—á–µ—Ç–∞")
print("="*50)
```
